```python
import os
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATA_PATH = "/Users/lukefoster/Documents/CompEtho/BM_GPU/results/cluster"
```

```python
min_DF = pd.read_csv(f"{DATA_PATH}/min_clusters.csv", index_col=0)
mid_DF = pd.read_csv(f"{DATA_PATH}/mid_clusters.csv", index_col=0)
maj_DF = pd.read_csv(f"{DATA_PATH}/maj_clusters.csv", index_col=0)
other_DF = pd.read_csv(f"{DATA_PATH}/other_clusters.csv", index_col=0)
```

```python
cluster_data = []

for cluster in range(min_DF.shape[1]-1):
    for proportion in min_DF[f"Cluster {cluster}"]:
        row = [cluster, "min", proportion]
        cluster_data.append(row)
        
for cluster in range(mid_DF.shape[1]-1):
    for proportion in mid_DF[f"Cluster {cluster}"]:
        row = [cluster, "mid", proportion]
        cluster_data.append(row)

for cluster in range(maj_DF.shape[1]-1):
    for proportion in maj_DF[f"Cluster {cluster}"]:
        row = [cluster, "maj", proportion]
        cluster_data.append(row)

'''
for cluster in range(other_DF.shape[1]-1):
    for proportion in other_DF[f"Cluster {cluster}"]:
        row = [cluster, "other", proportion]
        cluster_data.append(row)
'''
        
cluster_DF = pd.DataFrame(data=cluster_data)
cluster_DF = cluster_DF.rename(columns={0: "cluster", 1: "caste", 2: "proportion"})
```

```python
fig, ax = plt.subplots(figsize=(150,7))
ax = sns.boxplot(x="cluster", y="proportion", hue="caste", 
                 data=cluster_DF)
plt.savefig("all_clusters.png")
plt.show()
```
