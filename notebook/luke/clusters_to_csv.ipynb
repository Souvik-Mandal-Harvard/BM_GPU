```python
# establish path to cluster data
PROJECT_PATH = "/Users/lukefoster/Documents/CompEtho/BM_GPU"
CONFIG_PATH = f"{PROJECT_PATH}/config_ant.yaml"
```

```python
import sys
import pandas as pd
import numpy as np

sys.path.append(PROJECT_PATH)
from utils.data import Dataset
```

```python
# load all ant data
Data = Dataset(PROJECT_PATH, CONFIG_PATH)
Data.load_data()
```

```python
# save data information (including video directories) into INFO_values
INFO_values = Data.info_values
```

```python
# create empty list to store video folders in
folders = []

# fill folders list with all video directores within INFO_values
for folder in INFO_values:
    folders.append(folder['directory'])
```

```python
# create empty dictionary to store cluster data for each video file
cluster_data = {}

# fill cluster_data dictionary with cluster numpy data
for folder in folders:
    cluster_data[f"{folder.split('/')[2]}"] = np.load(f"{PROJECT_PATH}/{folder}/cluster.npy")
```

```python
cluster_data
```

```python
all_lengths = []

for key,value in cluster_data.items():
    all_lengths.append(len(value))

all_lengths.sort()
```

```python
arr = np.empty((len(cluster_data),all_lengths[-1]+1), dtype="object")


for i,dict in enumerate(cluster_data.items()):
    #row = np.full(all_lengths[-1]+1, -2, dtype="object")
    row = np.empty(all_lengths[-1]+1, dtype="object")
    
    row[0] = dict[0]
    
    j = 1
    for j,value in enumerate(dict[1]):
        row[j+1] = int(value)
    print(row)
        
    arr[i] = row
```

```python
# convert array into dataframe
DF = pd.DataFrame(arr)
  
# save the dataframe as a csv file
DF.to_csv("clusters.csv")
```
